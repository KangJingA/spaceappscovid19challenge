```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath = 'data/ExportSalesDataByCommodity.csv'
df = pd.read_csv(filepath, skiprows=[0,1,2,3,4,5])
df.columns
```

```python
cols = [0, 3]
df = df.drop(df.columns[cols], axis=1)
df = df.rename(columns={'Unnamed: 1': 'Commondity',
                        'Exports': 'Weekly Exports',
                        'Exports.1': 'Accum. Exports',
                        'Sales': 'Outstanding Sales',
                        'Sales.1': 'Gross Sales',
                        'Sales.2': 'Net Sales',
                        'Commitments': 'Total Commitments',
                        'Sales.3': 'Outstanding Sales_NMY',
                        'Sales.4': 'Net Sales_NMY' })
df.head()
```

# Data Preprocessing
